### Enemy 1 Comparison

```python


def process_df(df):
    df['run_id'] = df.index // 100
    df['generation'] = df.index % 100 
    
    grouped = df.groupby(['generation', 'run_id']).agg(
        mean_fitness_mean=('mean_fitness', 'mean'),
        best_fitness_mean=('best_fitness', 'mean')
    ).reset_index()
    
    overall = grouped.groupby('generation').agg(
        mean_fitness_avg=('mean_fitness_mean', 'mean'),
        mean_fitness_overall_std=('mean_fitness_mean', 'std'),
        best_fitness_avg=('best_fitness_mean', 'mean'),
        best_fitness_overall_std=('best_fitness_mean', 'std')
    ).reset_index()
    return overall


df1 = method1enemy1
df2 =method2enemy1

overall1 = process_df(df1)
overall2 = process_df(df2)

# Plotting
plt.figure(figsize=(12,6))

# Method 1
plt.plot(overall1['generation'], overall1['mean_fitness_avg'], 'b-', label='Method 1: Average Mean Fitness')
plt.fill_between(overall1['generation'], overall1['mean_fitness_avg'] - overall1['mean_fitness_overall_std'], overall1['mean_fitness_avg'] + overall1['mean_fitness_overall_std'], color='b', alpha=0.1)

plt.plot(overall1['generation'], overall1['best_fitness_avg'], 'r-', label='Method 1: Average Best Fitness')
plt.fill_between(overall1['generation'], overall1['best_fitness_avg'] - overall1['best_fitness_overall_std'], overall1['best_fitness_avg'] + overall1['best_fitness_overall_std'], color='r', alpha=0.1)

# Method 2
plt.plot(overall2['generation'], overall2['mean_fitness_avg'], 'g-', label='Method 2: Average Mean Fitness')
plt.fill_between(overall2['generation'], overall2['mean_fitness_avg'] - overall2['mean_fitness_overall_std'], overall2['mean_fitness_avg'] + overall2['mean_fitness_overall_std'], color='g', alpha=0.1)

plt.plot(overall2['generation'], overall2['best_fitness_avg'], 'c-', label='Method 2: Average Best Fitness')
plt.fill_between(overall2['generation'], overall2['best_fitness_avg'] - overall2['best_fitness_overall_std'], overall2['best_fitness_avg'] + overall2['best_fitness_overall_std'], color='c', alpha=0.1)
plt.legend(loc='best', fontsize='16')
plt.xlabel('Generation',fontsize='18')
plt.ylabel('Fitness', fontsize='18')
plt.title('Enemy 1 Method 1 and Method 2',fontsize='18')
plt.grid(True)
plt.show()
```
